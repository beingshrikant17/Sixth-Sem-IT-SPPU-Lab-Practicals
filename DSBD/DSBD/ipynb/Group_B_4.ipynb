```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

KS = {'color': ['blue', 'red', 'yellow']}
sns.FacetGrid(df, hue='Species', hue_kws=KS, height=5)\
    .map(plt.scatter, 'SepalLengthCm', 'SepalWidthCm')\
    .add_legend()
```

```python
sns.boxplot(x="Species",y="SepalWidthCm",data=df)
```

```python
ax=sns.stripplot(x='Species',y='SepalLengthCm',data=df,jitter=True,linewidth=2,edgecolor='pink')
```

```python
sns.violinplot(x='Species',y='SepalLengthCm',data=df,size=6,color="purple")
```

```python
sns.pairplot(df,hue='Species')
```
